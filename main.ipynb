```python
import pandas as pd
test = pd.read_csv('Test.csv', names=['Item_Identifier', 'Item_Weight', 'Item_Fat_Content', 'Item_Visibility', 'Item_Type', 'Item_MRP', 'Outlet_Identifier', 'Outlet_Establishment_Year', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type'])
print(test.head())
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

# scatter the sepal_length against the sepal_width
ax.scatter(test['Item_MRP'], test['Item_Visibility'])
# set a title and labels
ax.set_title('MRP angainst Visibility')
ax.set_xlabel('Item_MRP')
ax.set_ylabel('Item_Visibility')
```

```python
# create color dictionary
colors = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}
# create a figure and axis
fig, ax = plt.subplots()
# plot each data-point
for i in range(len(test['Outlet_Establishment_Year'])):
    ax.scatter(test['Outlet_Type'][i], test['Outlet_Establishment_Year'][i])
# set a title and labels
ax.set_title('Outlet Type Vs Items')
ax.set_xlabel('Outlet_Establishment_Year')
ax.set_ylabel('Outlet_Type')
```

```python
# get columns to plot
columns = test.columns.drop(['Item_Visibility'])
# create x data
x_data = range(0, test.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, test[column])
# set title and legend
ax.set_title('Iris Dataset')
ax.legend()
```
